```python
def counter(fname):
 
    # variable to store total word count
    num_words = 0
     
    # variable to store total line count
    num_lines = 0
     
    # variable to store total character count
    num_charc = 0
     
    # variable to store total space count
    num_spaces = 0
     
    
    with open(fname, 'r') as f:
        
        for line in f:
             
            # incrementing value of num_lines with each
            num_lines += 1
             
            # declaring a variable word and assigning its value as Y
            word = 'Y'
             
            
            for letter in line:
                 
                # condition to check that the encountered character
                # is not white space and a word
                if (letter != ' ' and word == 'Y'):
                     
                    # incrementing the word
                    num_words += 1
                     
                    # assigning value N to variable word because until
                    word = 'N'
                     
                # condition to check that the encountered character is a white space
                elif (letter == ' '):
                     
                    # incrementing the space
                    # count by 1
                    num_spaces += 1
                     
                    # assigning value Y to variable word because after
                    # white space a word is supposed to occur
                    word = 'Y'
                     
                # loop to iterate every character
                for i in letter:
                     
                    # condition to check white space
                    if(i !=" " and i !="\n"):
                         
                        # incrementing character
                        # count by 1
                        num_charc += 1
                         
    # printing total word count
    print("Number of words in text file: ",
          num_words)
     
    # printing total line count
    print("Number of lines in text file: ",
          num_lines)
     
    # printing total character count
    print('Number of characters in text file: ',
          num_charc)
     
    # printing total space count
    print('Number of spaces in text file: ',
          num_spaces)
     
# Driver Code:
if __name__ == '__main__':
    fname = 'file1.txt'
    try:
        counter(fname)
    except:
        print('File not found')
```

```python

# Open the file in read mode 
text = open("sample_text.txt", "r") 
  
# Create an empty dictionary 
d = dict() 
  
# Loop through each line of the file 
for line in text: 
    # Remove the leading spaces and newline character 
    line = line.strip() 
  
    # Convert the characters in line to 
    # lowercase to avoid case mismatch 
    line = line.lower() 
  
    # Split the line into words 
    words = line.split(" ") 
                         
  
    # Iterate over each word in line 
    for word in words: 
        # Check if the word is already in dictionary 
        if word in d: 
            # Increment count of word by 1 
            d[word] = d[word] + 1
        else: 
            # Add the word to dictionary with count 1 
            d[word] = 1
  
# Print the contents of dictionary 
for key in list(d.keys()): 
    print(key, ":", d[key]) 
```

# Question 2: Library Management System using Python OOP

```python
class Book:
    
    def __init__(self, title, author, isbn):
        self.title = title
        self.author = author
        self.isbn = isbn
        self.is_borrowed = False

    def borrow(self):
        if not self.is_borrowed:
            self.is_borrowed = True
            print(f"{self.title} by {self.author} is now borrowed.")
        else:
            print(f"{self.title} is already borrowed.")

    def return_book(self):
        if self.is_borrowed:
            self.is_borrowed = False
            print(f"{self.title} by {self.author} is returned.")
        else:
            print(f"{self.title} is not borrowed.")

class DigitalBook(Book):
    
    def __init__(self, title, author, isbn, file_format):
        super().__init__(title, author, isbn)
        self.file_format = file_format

    def borrow(self):
        super().borrow()
        print(f"The digital book '{self.title}' can be accessed online.")

class AudioBook(Book):
    
    def __init__(self, title, author, isbn, duration):
        super().__init__(title, author, isbn)
        self.duration = duration

    def borrow(self):
        super().borrow()
        print(f"The audiobook '{self.title}' is available for streaming.")

class User:
    
    def __init__(self, user_id, name):
        self.__user_id = user_id
        self.__name = name
        self.__borrowed_books = []

    def borrow_book(self, book):
        if book.is_borrowed:
            print(f"The book '{book.title}' is currently unavailable.")
        else:
            book.borrow()
            self.__borrowed_books.append(book)

    def return_book(self, book):
        if book in self.__borrowed_books:
            book.return_book()
            self.__borrowed_books.remove(book)
        else:
            print(f"The book '{book.title}' is not in the user's borrowed list.")

class Library:
    
    def __init__(self, name):
        self.name = name
        self.books = []
        self.users = []

    def add_book(self, book):
        self.books.append(book)
        print(f"Book '{book.title}' added to the library.")

    def register_user(self, user):
        self.users.append(user)
        print(f"User '{user._User__name}' registered with the library.")

    def book(self, user_id, isbn):
        user = self.find_user_by_id(user_id)
        book = self.find_book_by_isbn(isbn)

        if user and book and not book.is_borrowed:
            user.borrow_book(book)
        elif book and book.is_borrowed:
            print(f"Sorry, the book '{book.title}' is already borrowed.")
        else:
            print(f"User or book not found.")

    def receive_return(self, user_id, isbn):
        user = self.find_user_by_id(user_id)
        book = self.find_book_by_isbn(isbn)

        if user and book:
            user.return_book(book)
        else:
            print("User or book not found.")

    def find_user_by_id(self, user_id):
        for user in self.users:
            if user._User__user_id == user_id:
                return user
        return None

    def find_book_by_isbn(self, isbn):
        for book in self.books:
            if book.isbn == isbn:
                return book
        return None

# Test the Library Management System
if __name__ == "__main__":
    # Create library
    library = Library("Central Library")

    # Add books to library
    book1 = Book("bange darah", "Allama iqbal", "1234567890")
    ebook1 = DigitalBook("1930", "Tesere abwab us sarf", "0987654321", "PDF")
    audiobook1 = AudioBook("Moby Dick", "Herman Melville", "1122334455", 15.5)

    library.add_book(book1)
    library.add_book(ebook1)
    library.add_book(audiobook1)

    # Register users
    user1 = User("U001", "Alice")
    user2 = User("U002", "Bob")

    library.register_user(user1)
    library.register_user(user2)

    # User borrows and returns books
    library.book("U001", "1234567890")  # Alice borrows The Great Gatsby
    library.book("U002", "1234567890")  # Bob tries to borrow The Great Gatsby (already borrowed)
    library.book("U001", "0987654321")  # Alice borrows digital book 1984
    library.book("U002", "1122334455")  # Bob borrows audiobook Moby Dick

    # Users return books
    library.receive_return("U001", "1234567890")  # Alice returns The Great Gatsby
    library.receive_return("U001", "0987654321")  # Alice returns 1984
```
